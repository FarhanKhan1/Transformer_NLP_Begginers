```python
import pandas as pd
df = pd.read_csv('bbc_text_cls.csv')
```

```python
df.head(3)
```

```python
labels = list(set(df.labels))
```

```python
labels
```

```python
df.loc[4]
```

```python
zero_shot_classifier = pipeline('zero-shot-classification')
```

```python
zero_shot_classifier('i am in not in a good mood.',
                         candidate_labels = ['positive', 'negative']
                         )
```

```python
zero_shot_classifier(df.loc[4]['text'], 
                     candidate_labels = labels )
```
