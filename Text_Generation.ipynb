```python
from transformers import pipeline, set_seed
import textwrap
import numpy as np
import matplotlib.pyplot as plt
from pprint import pprint  
```

```python
!cat robert_frost.txt
```

```python
lines = [line.rstrip() for line in open('robert_frost.txt')]
line = [line for line in lines if len(line)>0]
```

```python
generator = pipeline('text-generation')
```

```python
  set_seed(1234)
```

```python
lines[0:3]
```

```python
generator(lines[0])
```

```python
pprint(_)
```

```python
pprint(generator(lines[0], max_length = 20))
```

```python
pprint(generator(lines[0], num_return_sequences = 4, max_length = 20))
```

```python
def wrap(x):
  return textwrap.fill(x, replace_whitespace=False, fix_sentence_endings=True)
```

```python
output = generator(lines[0], max_length=30)
wrap(output[0]['generated_text'])
```

```python
pre = 'Two roads diverged in a yellow wood, leaving an entire ditch open in\nfront of a small road.'
```

```python
output = generator(pre + '\n' + lines[2], max_length=60)
```

```python
wrap(output[0]['generated_text'])
```
